```python
import os
import glob
from datetime import datetime 
import pandas as pd
files = glob.glob('/home/user/BuildingsBenchNREL/BuildingsBenchData/BuildingsBench/LCL/*2012*.csv')
len(files)
```

```python
df_all = []
for f in files:        
    df = pd.read_csv(f)
    
    # if df.shape[0] != 8760:
    #     continue

    print(os.path.basename(f), df.shape)
        
    ## filename - MAC000005_clean=2012.csv
    name = os.path.basename(f).split('_')[0]
    year = os.path.basename(f).split('=')[1][:4]
    
    df.columns = ['timestamp', name + '_' + year]
    
    # # Order by timestamp
    df = df.sort_values(by='timestamp').round(4)
    # # Set timestamp as the index
    df.set_index('timestamp', inplace=True)
    df.index = pd.to_datetime(df.index, format='%Y-%m-%d %H:%M:%S')
    df_all.append(df)

combined = pd.concat(df_all, axis=1)
```

```python
print(combined.shape)
combined.to_csv('/home/user/BuildingsBenchNREL/BuildingsBenchData/BuildingsBench/LCL-csv/LCL-2012.csv')
combined.to_feather('/home/user/BuildingsBenchNREL/BuildingsBenchData/BuildingsBench/LCL-csv/LCL-2012.feather')
```
